## Define transformation pipeline

```python
crop_size = 225
upscale_factor = 2
```

```python
tfm = Compose([
        Scale(crop_size // upscale_factor),
        RandomCrop(crop_size // upscale_factor)])
```

```python
fonts = []
for d in os.listdir("./data"):
    if d.startswith(".") is False:
        fonts.append(d)
```

```python
fonts
```

## Run image transformation for all fonts

```python

for font in fonts:
    crops = []
    for f in os.listdir('./data/{}/'.format(font)):
#         print(f)
        if f.endswith("jpg") and len(f) > 5:


            # open file
            with Image.open('./data/{}/{}'.format(font, f)) as im:
                for i in range(10):
                    crop = tfm(im)
                    print('./data/{}/std/{}'.format(font, f))
                    crop.save('./data/{}/std/{}'.format(font, f))
                    crops.append(crop)
```
